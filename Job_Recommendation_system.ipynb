Importing Dependencies

```python
import numpy as np
import pandas as pd
import difflib 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data collection and pre-processing

```python
#loading the data from a csv file to a pandas file
jobs_data = pd.read_csv('/content/jobs.csv')
```

```python
jobs_data.head()
```

```python
jobs_data['index'] = [i for i in range(0,len(jobs_data.values))]
print(jobs_data.head()) 
```

```python
#selecting the relevant features 
selected_features= ['_id','profile','Location','Skills and Perks','Apply by Date']
print(selected_features)
```

```python
#replacing the null values
for feature in selected_features :
  jobs_data[feature]=jobs_data[feature].fillna('')
```

```python
#combining the 5 features
combined_features= jobs_data['_id']+jobs_data['profile']+jobs_data['Location']+jobs_data['Skills and Perks']+jobs_data['Stipend']
```

```python
print(combined_features)
```

```python
#converting the text data into feature vectors
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine Similarity


```python
#getting the similarity scores by using cosine similarity
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
#getting the name of the jobs 
jobs_name=input('Enter the job you are seeking: ') 
```

```python
#creating a list with all the jobs given in the dataset
list_of_all_profiles=jobs_data['profile'].tolist()
print(list_of_all_profiles)
```

```python
#finding the close match 
find_close_match=difflib.get_close_matches(jobs_name,list_of_all_profiles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
index_of_the_job=jobs_data[jobs_data.profile == close_match]["index"].values[0]
print(index_of_the_job)
```

```python
#getting the list of similar movies
similarity_score=list(enumerate(similarity[index_of_the_job]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_jobs= sorted(similarity_score,key= lambda x:x[1], reverse=True)
print(sorted_similar_jobs)
```

```python
i=1
for jobs in sorted_similar_jobs:
  index=jobs[0]
  profile_from_index= jobs_data[jobs_data.index==index]['profile'].values[0]
  if(i<30):
    print(i,'',profile_from_index)
    i+=1
```

```python
#Job Recommendation System:
jobs_name=input('Enter the job you are seeking: ')

list_of_all_profiles=jobs_data['profile'].tolist()

find_close_match=difflib.get_close_matches(jobs_name,list_of_all_profiles)

close_match=find_close_match[0]

index_of_the_job=jobs_data[jobs_data.profile == close_match]["index"].values[0]

similarity_score=list(enumerate(similarity[index_of_the_job]))

sorted_similar_jobs= sorted(similarity_score,key= lambda x:x[1], reverse=True)

print('Suggested Jobs are as follows:\n')

i=1
for jobs in sorted_similar_jobs:
  index=jobs[0]
  profile_from_index= jobs_data[jobs_data.index==index]['profile'].values[0]
  if(i<60):
    print(i,'',profile_from_index)
    i+=1
```
